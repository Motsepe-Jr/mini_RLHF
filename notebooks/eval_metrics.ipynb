```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Read the CSV file
df = pd.read_csv('../ppo_training_stats.csv')

# Display the first few rows of the dataframe
plt.figure(figsize=(12, 8))

df['ppo/loss/total'] = df['ppo/loss/total'].apply(lambda x: ast.literal_eval(x)[0])
# Convert the column to float type
df['ppo/loss/total'] = df['ppo/loss/total'].astype(float)
plt.plot(df['ppo/loss/total'], label='KL Divergence')
# plt.plot(df['ppo/policy/entropy'], label='Entropy')

plt.xlabel('Training Step')
plt.ylabel('Value')
plt.title('PPO Training Metrics')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
